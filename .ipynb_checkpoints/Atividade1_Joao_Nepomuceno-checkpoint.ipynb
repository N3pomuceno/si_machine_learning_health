# 1. 
Importarei as bibliotecas necessárias para produzir a atividade: Pandas para a leitura do arquivo e tratamento dos dados. Seaborn e matplotlib.pyplot para a produção de gráficos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2.
Ler o arquivo de dados .CSV.

```python
dados = pd.read_csv('dados/a1_breast_cancer.csv')
```

```python
# Por uma prática, vou expor o dataframe para checar se está tudo certo
dados.head()
```

Pelos insumos da Atividade 1, trabalharemos em cima de 6 colunas, e vou já separá-las para ter um maior foco já de início.

```python
dados = dados[['worst concave points', 'mean concavity','worst perimeter', 'worst radius', 'worst area', 'diagnosis']]
dados.head()
```

# 3.
Vamos trabalhar em cima dos gráficos de distribuição e de dispersão em cima das colunas selecionadas, e separando o diagnósticos em _malignant_ e _benign_. Farei isso separando cada coluna nos dois casos. **Sempre _benignant_ na esquerda e _malignant_ na direita**. Após isso, farei a utilização do método pairplot para fazer separadamente o benigno, maligno e a sobreposição dos dois para comparação.

Mas antes de começar a fazer os gráficos alterarei os displays dos gráficos para ficar mais visualmente agradável.

```python
dados_benign = dados[dados['diagnosis']=='benign']
dados_malignant = dados[dados['diagnosis']=='malignant']
```

## Histogramas:

```python
sns.set_style('darkgrid')
```

```python
x_max, x_min = max(max(dados_benign['worst concave points']), max(dados_malignant['worst concave points'])), min(min(dados_benign['worst concave points']), min(dados_malignant['worst concave points']))
print(x_max, x_min)
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(1,2,1)
ax.set_title('Distribuição de "Worst Concave points"',fontsize=18, loc='left')
sns.histplot(data=dados_benign['worst concave points'], color='blue')
plt.xlim(x_min, x_max)
plt.ylim(0, 60)
plt.subplot(1,2,2)
sns.histplot(data=dados_malignant['worst concave points'], color='red')
plt.xlim(x_min, x_max)
plt.ylim(0, 60)
ax=ax
```

```python
x_max, x_min = max(max(dados_benign['mean concavity']), max(dados_malignant['mean concavity'])), min(min(dados_benign['mean concavity']), min(dados_malignant['mean concavity']))
print(x_max, x_min)
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(1,2,1)
ax.set_title('Distribuição de "mean concavity"',fontsize=18, loc='left')
sns.histplot(data=dados_benign['mean concavity'], color='blue')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
plt.subplot(1,2,2)
sns.histplot(data=dados_malignant['mean concavity'],color='red')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
ax=ax
```

```python
x_max, x_min = max(max(dados_benign['worst perimeter']), max(dados_malignant['worst perimeter'])), min(min(dados_benign['worst perimeter']), min(dados_malignant['worst perimeter']))
print(x_max, x_min)
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(1,2,1)
ax.set_title('Distribuição de "Worst Perimeter"',fontsize=18, loc='left')
sns.histplot(data=dados_benign['worst perimeter'], color='blue')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
plt.subplot(1,2,2)
sns.histplot(data=dados_malignant['worst perimeter'],color='red')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
ax=ax
```

```python
x_max, x_min = max(max(dados_benign['worst radius']), max(dados_malignant['worst radius'])), min(min(dados_benign['worst radius']), min(dados_malignant['worst radius']))
print(x_max, x_min)
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(1,2,1)
ax.set_title('Distribuição de "Worst Radius"',fontsize=18, loc='left')
sns.histplot(data=dados_benign['worst radius'], color='blue')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
plt.subplot(1,2,2)
sns.histplot(data=dados_malignant['worst radius'],color='red')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
ax=ax
```

```python
x_max, x_min = max(max(dados_benign['worst area']), max(dados_malignant['worst area'])), min(min(dados_benign['worst area']), min(dados_malignant['worst area']))
print(x_max, x_min)
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(1,2,1)
ax.set_title('Distribuição de "Worst Area"',fontsize=18, loc='left')
sns.histplot(data=dados_benign['worst area'], color='blue')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
plt.subplot(1,2,2)
sns.histplot(data=dados_malignant['worst area'],color='red')
plt.xlim(x_min, x_max)
plt.ylim(0, 65)
ax=ax
```

## Gráficos de Dispersão:

### Caso _Benigno_:

```python
sns.pairplot(dados_benign[['worst concave points', 'mean concavity',
                                                  'worst perimeter', 'worst radius', 
                                                  'worst area']])
```

### Caso _Maligno_:

```python
sns.pairplot(dados_malignant[['worst concave points', 'mean concavity',
                                                     'worst perimeter', 'worst radius', 
                                                     'worst area']])
```

### Sobreposto:

```python
colors = {
    'benign': 'blue',
    'malignant': 'red'
}
ax = sns.pairplot(data=dados, hue='diagnosis',palette=colors)
ax=ax
```

Observação: O interessante dessa sobreposição dos gráficos de sobreposição é que os gráficos que estão na diagonal principal dessa 'matriz', mostram as distribuições que vimos de cada uma das colunas! E mostra que uma coisa que pode ser passada quando olhando as distribuições uma a uma: Os valores do eixo vertical dos diagnósticos 'benigno' são maiores que os dos diagnóstico 'maligno'.
